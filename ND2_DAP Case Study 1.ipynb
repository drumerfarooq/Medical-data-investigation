```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display as disp

%matplotlib inline

red_df = pd.read_csv('winequality-red.csv')
disp(red_df.head())
# We can see below that the values are not separated by commas but semi-colons
# We can fix this by using the sep parameter

```

```python
# We can see above that the values are not separated by commas but semi-colons
# We can fix this by using the sep parameter
red_df = pd.read_csv('winequality-red.csv', sep = ';')
disp(red_df.head())
disp(red_df.describe())
disp(red_df.info())
white_df = pd.read_csv('winequality-white.csv', sep = ';')
disp(white_df.head())
disp(white_df.describe())
disp(white_df.info())
```

```python
# TASK: combine both datasets into a single dataframe
# create a numpy array for a new column COLOR (for each dataframe)
# insert the new column into each dataframe 
# then append the two dataframes
```

```python
# creating the numpy arrays

red_array = np.full(red_df.shape[0], 'R')
print(red_array.shape)
white_array = np.full(white_df.shape[0], 'W')
print(white_array.shape)

# Adding the color columns to both data frames
red_df['color'] = red_array
disp(red_df.head())
disp(red_df.tail())
white_df['color'] = white_array
disp(white_df.head())
disp(white_df.tail())

# This adds the column to the end lets pop it and put it as the first column
red_df.pop('color')
disp(red_df.head())
white_df.pop('color')
disp(white_df.head())

# Same can be done using drop method (drop can aslo be used to delete rows unlike pop)
red_df['color'] = red_array
red_df.drop(['color'],axis=1, inplace = True) # need to use inplace parameter here to save changes
disp(red_df.head())
white_df['color'] = white_array
white_df.drop(['color'],axis=1, inplace = True)
disp(white_df.head())

# Now we use the insert method to place the new column at location of our choice
red_df.insert(0, 'color', red_array)
disp(red_df.head())
white_df.insert(0, 'color', white_array)
disp(white_df.head())
```

```python
# finally we have the new dataframes with the added color column and we can simply append them

disp(red_df.head())
disp(white_df.head())

combined_df = red_df.append(white_df) 
# make sure all column names are same either manually or by code
# If there are mismatching column names the column will be filled with NaNs
disp(combined_df.head())
disp(combined_df.tail())
disp(combined_df.describe())
disp(combined_df.info())

# You can notice that the indices are still the ones from the original dataframes
# This can be fixed by resetting the indices

combined_df.index = np.arange(combined_df.shape[0])
disp(combined_df.head())
disp(combined_df.tail())

# to save this combined dataframe as a new csv

combined_df.to_csv('combinedwinedata.csv', index = False) # index = false to avoid creating an unnamed column of indices 
```

```python
# Drawing conclusions

# Q1: Is a certain type of wine (red or white) associated with higher quality?

# Q2: What level of acidity (pH value) receives the highest average rating?
```

```python
# Q1: Is a certain type of wine (red or white) associated with higher quality?

disp(combined_df.head())
disp(combined_df.groupby('color',as_index = False)['quality'].mean())
combined_df.groupby('color',as_index = False)['quality'].mean().to_csv('series_to_csv.csv', index = False)
# white wine has higher average quality

disp(combined_df.describe()) # for Q2
```

```python
# Q2: What level of acidity (pH value) receives the highest average rating?

# To create some ranges for acidity levels (lets say 4) we need to use the pandas cut function

# View the min, 25%, 50%, 75%, max pH values with Pandas describe

# Bin edges that will be used to "cut" the data into groups
bin_edges = [ 2.72,3.11 ,3.21 ,3.32 ,4.01 ] # Fill in this list with five values you just found

# Labels for the four acidity level groups
bin_names = ['(0 - 25)%' ,'(25 - 50)%' ,'(50 - 75)%' ,'(75 - 100)%' ] # Name each acidity level category

# Creates acidity_levels column
combined_df['acidity_levels'] = pd.cut(combined_df['pH'], bin_edges, labels=bin_names)

# Checks for successful creation of this column
disp(combined_df.head())

# Now we just get the mean quality for the ranges 
disp(combined_df.groupby('acidity_levels')['quality'].mean())
combined_df.to_csv('combinedwinedata1.1.csv', index=False)

# just for pratice
pH_range_practice = pd.cut(combined_df['pH'], bins = 4)
disp(pH_range_practice)
pH_range_practice = pd.cut(combined_df['pH'], bins = 4, labels = False)
disp(pH_range_practice)
```

```python
# using queries

# Q1: Do wines with higher alcoholic content receive better ratings?

# Q2: Do sweeter wines (more residual sugar) receive better ratings?
```

```python
# Q1: Do wines with higher alcoholic content receive better ratings?

# getting means of all qualities with alcohol content higher and lower than the median alcohol value (can also be mean etc) 
disp(combined_df[combined_df['alcohol'] > combined_df['alcohol'].median()]['quality'].mean())
disp(combined_df[combined_df['alcohol'] < combined_df['alcohol'].median()]['quality'].mean())

# It is clear that higher alcohol content gives better quality
```

```python
# Q2: Do sweeter wines (more residual sugar) receive better ratings?

# getting means of all qualities with residual sugar content higher and lower than the mean residual sugar value
disp(combined_df[combined_df['residual sugar'] > combined_df['residual sugar'].mean()]['quality'].mean())
disp(combined_df[combined_df['residual sugar'] < combined_df['residual sugar'].mean()]['quality'].mean())

# It is clear that lower residual sugar content gives better quality
```
